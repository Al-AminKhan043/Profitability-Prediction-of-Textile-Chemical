```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'SB International', 'MultiTrade', 'Pacific Trading'
sizes = [65, 25, 10]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)  # explode 1st slice (SB International)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',  startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'SB International', 'MultiTrade', 'Pacific Trading'
sizes = [65, 25, 10]
colors = ['gold', 'yellowgreen', 'lightcoral']

# Plot
plt.figure(figsize=(6, 6))  # Set figure size for publication quality
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.savefig("pie_chart_elsevier_format.png", format="png", dpi=300, bbox_inches='tight')

plt.show()
```
